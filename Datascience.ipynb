```python

def clean_extracted_table(df):


    df = df.dropna(axis=1, how="all")

    df = df.iloc[2:].reset_index(drop=True)

    df = df.replace({r"Â": "", r"\s+": " "}, regex=True)

    df.iloc[0] = df.iloc[0].fillna(method="ffill")


    df.columns = df.iloc[0]
    df = df[1:].reset_index(drop=True)

    return df


table_9_cleaned = clean_extracted_table(raw_table_9)
table_12_cleaned = clean_extracted_table(raw_table_12)
table_62_cleaned = clean_extracted_table(raw_table_62)


table_9_cleaned.head(), table_12_cleaned.head(), table_62_cleaned.head()
```

```python

def restructure_table(df):

    df.columns = df.iloc[0]
    df = df[1:].reset_index(drop=True)

    df = df.loc[:, ~df.columns.duplicated()]

    df.columns = [str(col).strip() for col in df.columns]

    return df

table_9_final = restructure_table(table_9_cleaned)
table_12_final = restructure_table(table_12_cleaned)
table_62_final = restructure_table(table_62_cleaned)

table_9_final.head(), table_12_final.head(), table_62_final.head()
```

```python
import pandas as pd


file_path = "/content/merged_output(3).xlsx"

df = None
try:
    df = pd.read_excel(file_path, header=None, engine="openpyxl")
except Exception as e:
    print("Error loading file:", e)

if df is not None:

    filename = "table9"
    label = "Revenues ($ millions) (2)"
    data = []


    for i in range(0, len(df), 4):
        if i + 4 >= len(df):
            break

        tabletitle = df.iloc[i, 0]
        for col in range(1, df.shape[1]):
            try:
                column_header = df.iloc[i, col]
                value_2022 = df.iloc[i + 1, col] if i + 1 < len(df) else None
                value_2021 = df.iloc[i + 2, col] if i + 2 < len(df) else None
                change_fav = df.iloc[i + 3, col] if i + 3 < len(df) else None
                change_currency_fav = df.iloc[i + 4, col] if i + 4 < len(df) else None

                data.append([filename, label, tabletitle, column_header, value_2022])
                data.append([filename, label, tabletitle, column_header, value_2021])
                data.append([filename, label, tabletitle, "% Change Fav (Unfav)", change_fav])
                data.append([filename, label, tabletitle, "% Change at constant currency Fav (Unfav)", change_currency_fav])
            except IndexError:
                print(f"Skipping incomplete data at row {i} and column {col}")
                continue


    normalized_df = pd.DataFrame(data, columns=["filename", "label", "tabletitle", "column header", "value"])


    normalized_df.to_excel("transformed_data.xlsx", index=False)

    print("Data transformation complete. Saved as transformed_data.xlsx")
```
